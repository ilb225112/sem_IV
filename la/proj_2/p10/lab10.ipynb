```python
#Q3. Plot the triangle’s vertices and set the number of iterations; store points in x and plot them after all iterations.
import numpy as np 
import matplotlib.pyplot as plt 

# Define the vertices of the triangle 
t = np.linspace(0, 2 * np.pi, 4)[:-1]  # Generate t and remove the last element 
v = np.array([np.cos(t), np.sin(t)])  # 2×3 matrix of triangle vertices

# Define the linear transformation matrix 
T = np.array([[0.5, 0], [0, 0.5]]) 

# Define the starting point 
x = np.random.rand(2, 1) - 0.5 

# Number of iterations 
Num = 10000 

# Create an array to store all points 
points = np.zeros((2, Num)) 
current_point = x 

# Iterative process 
for i in range(Num): 
    points[:, i] = current_point.flatten()  # Store the current point
    vertex = v[:, np.random.choice(3)].reshape(2, 1)  # Choose random vertex
    current_point = T @ (current_point + vertex)  # Update the point

# Plot the vertices of the triangle 
plt.plot(v[0, :], v[1, :], 'k*', label='Vertices') 

# Plot the starting point 
plt.plot(x[0, 0], x[1, 0], 'b.', label='Starting Point') 

# Plot all the points obtained during the iterations 
plt.plot(points[0, :], points[1, :], 'r.', markersize=1, label='Points') 

plt.axis('equal') 
plt.legend() 
plt.title("Sierpinski Triangle Iterations")
plt.show()
```

```python
#Q4. Use random vertex selection and the transformation matrix T to iteratively generate and plot points in x for the Sierpinski triangle.
import numpy as np 
import matplotlib.pyplot as plt 

# Define the vertices of the triangle
t = np.linspace(0, 2 * np.pi, 4)[:-1]  # Generate t and remove the last element 
v = np.array([np.cos(t), np.sin(t)])  # 2×3 matrix of triangle vertices

# Define the linear transformation matrix
T = np.array([[0.5, 0], [0, 0.5]])

# Define the starting point
x = np.random.rand(2, 1) - 0.5

# Number of iterations
Num = 10000

# Create an array to store all points
points = np.zeros((2, Num + 1))
points[:, 0] = x.flatten()

# Iterative process
for j in range(Num):
    k = np.random.randint(0, 3)  # Random integer from 0 to 2
    current_point = points[:, j]  # Get the current point
    transformed_point = T @ (current_point - v[:, k]) + v[:, k]
    points[:, j + 1] = transformed_point

# Plot the points
plt.plot(points[0, :], points[1, :], 'b.', markersize=1)
plt.axis('equal')
plt.title('Sierpinski Triangle')
plt.show()
```

```python
#Q9. Square with Vertex Preference — Avoid Repeating the Same Vertex
# Define the vertices of the square using complex exponentials 
t = np.linspace(0, 2 * np.pi, 5)[:-1] 
v = np.exp(1j * t)  # Complex exponential for the vertices 

# Define the linear transformation matrix 
T = np.array([[0.5, 0], [0, 0.5]]) 

# Define the starting point 
x = np.random.rand(2, 1) - 0.5 

# Number of iterations 
Num = 5000 

# Create an array to store all points 
points = np.zeros((2, Num + 1)) 

# Initial point 
points[:, 0] = x.flatten() 

# Variable to keep track of the previous vertex index 
k1 = -1

# Iterative process 
for j in range(Num): 
    k = np.random.randint(0, 4) 
    if k >= k1: 
        k = (k + 1) % 4  # Ensure the same vertex is not selected twice in a row
     
    w = np.array([np.real(v[k]), np.imag(v[k])]) 
    current_point = points[:, j] 
    transformed_point = T @ (current_point - w) + w 
    points[:, j + 1] = transformed_point 
    k1 = k 
 
# Plot the result 
plt.figure() 
plt.plot(np.real(v), np.imag(v), 'k*', label='Vertices') 
plt.plot(points[0, :], points[1, :], 'b.', markersize=1, label='Fractal Points') 
plt.axis('equal') 
plt.title('Fractal with Vertex Preference') 
plt.legend() 
plt.show()
```

```python
#Q10. Square with Vertex Preference (No Opposite Vertex Allowed)

# Define the vertices of the square using complex exponentials 
t = np.linspace(0, 2 * np.pi, 5)[:-1] 
v = np.exp(1j * t)  # Complex exponential for the vertices 
 
# Define the linear transformation matrix 
T = np.array([[0.5, 0], [0, 0.5]]) 
 
# Define the starting point 
x = np.random.rand(2, 1) - 0.5 
 
# Number of iterations 
Num = 5000 
 
# Create an array to store all points 
points = np.zeros((2, Num + 1)) 
 
# Initial point 
points[:, 0] = x.flatten() 
 
# Track the previous vertex index 
k1 = 0 
w = np.array([np.real(v[k1]), np.imag(v[k1])]) 
 
# Iterative process 
for j in range(Num): 
    k = np.random.randint(0, 4)  # Random integer from 0 to 3
     
    # Avoid opposite vertex
    if (k != (k1 + 2) % 4) and (k1 != (k + 2) % 4): 
        w = np.array([np.real(v[k]), np.imag(v[k])]) 
        transformed_point = T @ (points[:, j] - w) + w 
        points[:, j + 1] = transformed_point 
        k1 = k 
    else: 
        points[:, j + 1] = points[:, j]  # No transformation if opposite vertex
 
# Plot the points and vertices
plt.figure() 
plt.plot(np.real(v), np.imag(v), 'k*', label='Vertices') 
plt.plot(points[0, :], points[1, :], 'b.', markersize=1, label='Fractal Points') 
plt.axis('equal') 
plt.title('Fractal with Vertex Preference (No Opposite Vertex)') 
plt.legend() 
plt.show()
```

```python
 #Q11. Barnsley Fern
# Define the transformation matrices and translation vectors 
T1 = np.array([[0.85, 0.04], [-0.04, 0.85]]) 
T2 = np.array([[-0.15, 0.28], [0.26, 0.24]]) 
T3 = np.array([[0.2, -0.26], [0.23, 0.22]]) 
T4 = np.array([[0, 0], [0, 0.16]]) 
Q1 = np.array([0, 1.64]) 
Q2 = np.array([-0.028, 1.05]) 
Q3 = np.array([0, 1.6]) 
Q4 = np.array([0, 0]) 

# Associated probabilities
P1, P2, P3, P4 = 0.85, 0.07, 0.07, 0.01

# Number of points
Num = 15000

# Initialize the starting point
x = np.zeros((2, Num + 1)) 
x[:, 0] = np.random.rand(2)  # Starting point

# Plot the initial point
plt.figure() 
plt.plot(x[0, 0], x[1, 0], 'b.') 

# Iterative process to generate the fern 
for j in range(Num): 
    r = np.random.rand() 
    if r <= P1: 
        x[:, j + 1] = T1 @ x[:, j] + Q1 
    elif r <= P1 + P2: 
        x[:, j + 1] = T2 @ x[:, j] + Q2 
    elif r <= P1 + P2 + P3: 
        x[:, j + 1] = T3 @ x[:, j] + Q3 
    else: 
        x[:, j + 1] = T4 @ x[:, j] + Q4 

# Plot the full fractal
plt.plot(x[0, :], x[1, :], 'b.', markersize=1) 
plt.axis('equal') 
plt.title('Barnsley Fern') 
plt.show()
```

```python
#Q12. Modify the Sierpinski triangle code to generate a fractal based on a regular hexagon using the transformation matrix
import numpy as np
import matplotlib.pyplot as plt

# Define the vertices of a regular hexagon
t = np.linspace(0, 2 * np.pi, 7)[:-1]  # Generate 6 vertices and remove the last one
v = np.array([np.cos(t), np.sin(t)])  # Shape: (2, 6)

# Define the linear transformation matrix
T = np.array([[0.25, 0], [0, 0.25]])  # Shrinks towards vertex

# Number of iterations
Num = 15000

# Initialize the points array
points = np.zeros((2, Num + 1))
points[:, 0] = np.random.rand(2) - 0.5  # Start near center

# Iterative process to generate the fractal
for j in range(Num):
    k = np.random.randint(0, 6)  # Choose one of the 6 vertices
    w = v[:, k]                  # Chosen vertex
    points[:, j + 1] = T @ (points[:, j] - w) + w  # Transform toward vertex

# Plot the result
plt.figure()
plt.plot(v[0, :], v[1, :], 'k*', label='Vertices')  # Hexagon vertices
plt.plot(points[0, :], points[1, :], 'b.', markersize=1, label='Fractal Points')  # Fractal points
plt.axis('equal')
plt.title('Fractal Based on Regular Hexagon')
plt.legend()
plt.show()
```

```python
#Q13. Create a Chaos Game using a regular pentagon, transformation matrix  and the rule that the next vertex must not be the same as the previous one.
import numpy as np
import matplotlib.pyplot as plt

# Define the vertices of a regular pentagon
t = np.linspace(0, 2 * np.pi, 6)[:-1]  # 5 vertices
v = np.array([np.cos(t), np.sin(t)])

# Define the linear transformation matrix
T = np.array([[2/5, 0], [0, 2/5]])

# Define the number of iterations
Num = 10000

# Initialize the starting point
x = np.zeros((2, Num + 1))
x[:, 0] = np.random.rand(2) - 0.5

# Store all points
points = np.zeros((2, Num + 1))
points[:, 0] = x[:, 0]

# Previous vertex index
prev_vertex_index = np.random.randint(0, 5)

# Iterative process
for j in range(Num):
    current_vertex_index = prev_vertex_index
    while current_vertex_index == prev_vertex_index:
        current_vertex_index = np.random.randint(0, 5)
    w = v[:, current_vertex_index]
    points[:, j + 1] = T @ (points[:, j] - w) + w
    prev_vertex_index = current_vertex_index

# Plot the vertices and the points
plt.figure()
plt.plot(v[0, :], v[1, :], 'k*', label='Vertices')  # Pentagon vertices
plt.plot(points[0, :], points[1, :], 'b.', markersize=1, label='Fractal Points')  # Fractal
plt.axis('equal')
plt.title('Fractal Based on Regular Pentagon with Vertex Restriction')
plt.legend()
plt.show()
```
